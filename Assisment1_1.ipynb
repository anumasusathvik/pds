```python
import pandas as pd
```

```python
raw_data = pd.read_csv("StudentsPerformance.csv")
```

```python
raw_data.head()
```

```python
raw_data.tail()
```

```python
raw_data.isnull()
```

```python
raw_data.isnull().sum()
```

```python
raw_data.drop_duplicates(inplace=True)
```

```python
raw_data['Average Score'] = (raw_data['math score']+raw_data['reading score']+raw_data['writing score'])/3
```

```python
raw_data.to_csv("cleaned_-data.csv" , index=False)
```

```python
clean_data= pd.read_csv("cleaned_-data.csv")
```

```python
clean_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
scores = sns.scatterplot(data = clean_data , x='race/ethnicity' , y = 'Average Score')
graph=scores.get_figure()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='gender', y='math score', data=clean_data, palette='Set3')
plt.xlabel('gender')
plt.ylabel('math Score')
plt.title('Box Plot of gender vs math Score')
plt.xticks(rotation=45, horizontalalignment='right')
```

```python
ax = sns.countplot(y="race/ethnicity" , hue="test preparation course" , data = clean_data)
graph = ax.get_figure()
```

```python
sns.jointplot(x='math score',y='reading score' , hue='gender' , data=clean_data)
```

```python

import plotly.express as px
fig = px.pie(clean_data, values="Average Score", names="race/ethnicity",
             title='score', hover_data=['gender'])
fig.show()
```
